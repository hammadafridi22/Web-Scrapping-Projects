```python
import time

prev_count = 0

while True:
    rows = driver.find_elements(By.TAG_NAME, "tr")
    curr_count = len(rows)

    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(2)

    if curr_count == prev_count:
        break

    prev_count = curr_count

print("Scrolling finished")
print("Total rows loaded:", curr_count)
```

```python
crypto_data = []

for row in rows[1:]:
    tds = row.find_elements(By.TAG_NAME, "td")
    if len(tds) < 9:
        continue

    rank = tds[1].text.strip()
    name = tds[2].text.split("\n")[0].strip()
    price = tds[3].text.strip()

    crypto_data.append([rank,
        name,
        price,
        hour_change,
        day_change,
        seven_days_change,
        market_cap,
        volume,
        circulating_supply])
```

```python
import pandas as pd

headers = ["Rank", "Name", "Price" , "1h %", "24h %", "7d %", "Market Cap", "Volume", "Circulating Supply"]

df = pd.DataFrame(crypto_data, columns=headers)
df.to_csv("Coin_Market.csv", index=False)

print("CSV file created successfully!")
```

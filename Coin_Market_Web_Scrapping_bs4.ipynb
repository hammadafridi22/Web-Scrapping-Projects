```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://coinmarketcap.com/'
request = requests.get(url)
```

```python
soup = BeautifulSoup(request.content,'html.parser') # urlib
soup
```

```python
table = soup.find('table' , class_ = 'sc-7e3c705d-3 keBvNC cmc-table')
table
```

```python
index_bitcoin = table.find_all('tr')
index_bitcoin
```

```python
table_data = index_bitcoin[2].find_all('td')
table_data
```

```python
rank = table_data[1].text.strip()
print(rank)
```

```python
for all_rank in index_bitcoin[2:]:
    rank_off = all_rank.find_all('td')
    rank = rank_off[1].text
    name = rank_off[2].text
    price = rank_off[3].text
    hourly_change = rank_off[4].text

    print(rank,name,price,hourly_change)



```
